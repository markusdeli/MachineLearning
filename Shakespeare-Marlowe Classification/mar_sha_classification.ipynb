```python
def find(word, ch):
    return [i for i, ltr in enumerate(word) if ltr == ch]
```

```python
def readTextLineForLine(doc):
  doc = nltk.word_tokenize(doc.read().lower())
  result = []
  tmp = []
  for word in doc:
    if word == '.' or word == '!' or word == '?' or word == ':':
      if len(tmp) != 0:
        result.append(tmp)
      tmp = []
    elif  word != ',' and word != ';'   and word != '"' and word != '\'' and word != 's' and word != '\'s' and word != '’' and word != '(' and word != ')' and word != '[' and word != ']' and word != '\'t' and word != '``' and word !='\'d' and word != '--' and word != '\n' and word != '' and word != "''" and word != '&' and any(chr.isdigit() for chr in word) == False:
      if word == '\'ll':
        word = 'will'
      if '—' in word:
        word = word.replace("—", "")
      if '-' in word:
        word = word[find(word, '-')[0] + 1:]
      tmp.append(word)
  return result
```

```python
def getWordcount(doc, word):
  count = 0
  for sentence in doc:
    for token in sentence:
      if token == word:
        count += 1
  return count
```

```python
def getX_XAnd(doc, word):
  count = 0
  occ = 0 
  for s in doc:
    count2 = 0
    for w in s:
      if count2 + 2 < len(doc[count]):
        if w == word:
          if doc[count][count2 + 2] == 'and':
            occ += 1
      count2 += 1
    count += 1
  return occ
```

```python
def getT10(doc):
  count = 0
  count += getWordcount(doc, 'but')
  count += getWordcount(doc, 'by')
  count += getWordcount(doc, 'for')
  count += getWordcount(doc, 'no')
  count += getWordcount(doc, 'not')
  count += getWordcount(doc, 'so')
  count += getWordcount(doc, 'that')
  count += getWordcount(doc, 'the')
  count += getWordcount(doc, 'to')
  count += getWordcount(doc, 'with')
  return count
```

```python
import statistics
def getAvgWordLength(doc):
  lengthCount = 0
  wordCount = 0
  length = []
  for s in doc:
    for word in s:
      lengthCount += len(word)
      wordCount += 1
      length.append(len(word))
  return lengthCount/wordCount, statistics.median(length)
```

```python
def getAvgSentenceLength(doc):
  lengthCount = 0
  length = []
  sentenceCount = 0
  for s in doc:
    tmp = 0
    for word in s:
      lengthCount += 1
      tmp += 1
    sentenceCount += 1
    length.append(tmp)
  return lengthCount/sentenceCount, statistics.median(length)
```

```python
def getVocSize(doc):
  liste = []
  for s in doc:
    for w in s:
      liste.append(w)
  return len(list(set(liste)))
```

```python
from nltk.corpus import stopwords 
nltk.download('stopwords')
def stopwordRatio(doc):
  stopWords = set(stopwords.words('english'))
  count = 0
  stopCount = 0
  tmp = []
  for s in doc:
    tmpCount = 0
    for word in s:
      count += 1
      if word in stopWords:
        stopCount += 1
        tmpCount += 1
    tmp.append(tmpCount)
  return stopCount/count, statistics.median(tmp) 
```

```python

def stem(docs):
  result = []
  for i in range(len(docs)):
    doc = []
    for s in docs[i][0]:
      sen = []
      for w in s:
        if '-' in w:
          w = w[find(w, '-')[0] + 1:]
        sen.append(stemmer.stem(w))
      doc.append(sen)
    result.append(doc)
  return result
```

```python
def stem2 (doc):
  result = []
  for s in doc:
    tmp = []
    for w in s:
      if '-' in w:
        w= w[find(w, '-')[0] + 1:]
      tmp.append(stemmer.stem(w))
    result.append(tmp)
  return result
```

```python
def flatList(docs):
  result = []
  for doc in docs:
    for s in doc:
      for w in s:
        result.append(w)
  return result
```

```python
def getAllWordsCount(docs):
  doc = flatList(docs)
  words = list(set(doc))
  result = []
  tmp = []
  for word in words:
    count = 0
    for w in doc:
      if w == word:
        count += 1
    tmp.append([count, word])
  tmp.sort(reverse = True)
  return tmp
```

```python
from collections import Counter

def getTF(doc):
  cpy = []
  for s in doc:
    for w in s:
      cpy.append(w)
  counts = Counter(cpy)
  result = []
  tmp = 0
  for i in range(len(allWords)):
    b = True
    for key in counts:
      if key == allWords[i][1]:
        result.append(counts[key])
        b = False
      if i == 0:
        tmp += counts[key]
    if b:
      result.append(0)
  for i in range(len(result)):
    if result[i] != 0:
      result[i] = result[i]/tmp
  return result
```

```python
def getSplit(doc, name, author):
  result = []
  count = 0
  tmp = []
  for s in doc:
    for w in s:
      count += 1
    tmp.append(s)
    if count >= 1000:
      count = 0
      result.append([tmp, author, name])
      tmp = []
  return result
```

```python
def getData(docs):
  liste = []
  
  for i in range(len(docs)):
    for z in range(len(getSplit(docs[i][0], docs[i][2], docs[i][1]))):  
      doc = getSplit(docs[i][0], docs[i][2], docs[i][1])
      doc = doc[z]
      t10 = getT10(doc[0])
      of = getWordcount(doc[0], 'of')
      the =  getWordcount(doc[0], 'the')
      meanWordLen, medianWordLen = getAvgWordLength(doc[0])
      meanSenLen, medianSenLen = getAvgSentenceLength(doc[0])
      meanStopwords, medianStopwords = stopwordRatio(doc[0])
      tmp = [getWordcount(doc[0], 'no') / t10, getX_XAnd(doc[0], 'of') / of, getWordcount(doc[0], 'so') / t10, getX_XAnd(doc[0], 'the') / the, getWordcount(doc[0], 'with') / t10, meanWordLen, medianWordLen, meanSenLen, medianSenLen, meanStopwords, medianStopwords, getVocSize(doc[0])/len(doc[0]), doc[1], doc[2] ]
      tf = getTF(stem2(doc[0]))
      lenTF = len(tf)
      for y in range(lenTF):
        tmp.append(tf[y])
      liste.append(tmp)
  return liste
```

```python
def getPredLists(preds, nameCounts):
  index = 0
  result = []
  for count in nameCounts:
    tmp = []
    for i in range(count):
      index += 1
      tmp.append(preds[index])
    result.append(tmp)
  res = []
  for liste in result:
    con = 0
    index = 0
    for i in range(len(liste)):
      index += 1
      con += liste[i]
    res.append(round(con/index, 3))
  return res
```

```python
def numbToAuth(preds):
  result = []
  for element in preds:
    if element > 1.5:
      author = '    S'
    else:
      author = '    M'
    result.append(str(element) + author)
  return result
```
